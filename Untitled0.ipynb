Desafio 1: Proporção de inscritos por idade

```
ageUniqueValues=data['NU_IDADE'].value_counts()
totalValues=data['NU_IDADE'].count()

agePercentage=(ageUniqueValues/totalValues*100).round(decimals=3).sort_index()
print(agePercentage.astype(str) + '%') 
```

Desafio 4: Histograma das idades dos treineiros e não treineiros

```
import matplotlib.pyplot as plt

data.query('IN_TREINEIRO==1')['NU_IDADE'].hist(bins=20, figsize=(10,8))

data.query('IN_TREINEIRO!=1')['NU_IDADE'].hist(bins=20, figsize=(10,8), alpha=0.6)

pl.title('Distribuição comparativa entre treineiros e não treineiros', fontsize=16)
plt.legend(['Treineiro', "Não treineiro"],fontsize=16, frameon=False, loc='center right')

pass
```

Desafio 5: Histograma comparando as notas da provas de Inglês e Espanhol

```
data.query('TP_LINGUA==0')['NU_NOTA_LC'].hist(bins=20, figsize=(10,6))
data.query('TP_LINGUA==1')['NU_NOTA_LC'].hist(bins=20, figsize=(10,6), alpha=0.6)

pl.title('Distribuição nota da prova de línguas entre inglês e espanhol', fontsize=16)
plt.legend(['Inglês', 'Espanhol'], frameon=False, fontsize=14, loc='upper left')
```
